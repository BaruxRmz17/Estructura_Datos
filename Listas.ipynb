# **Listas** 

Las listas simplemente enlazadas
- Cada nodo apunta solo al siguinte 
- El recorrido es unidireccional 
- El ultimo nodo a punta a None



```python
#Listas Ejercicio 1
class Nodo:
    def __init__(self,data):
        self.data=data
        self.sig = None


def imprimirLista(head):
    actualNodo = head
    while actualNodo != None:
        print(actualNodo.data, end=" -> ")
        actualNodo = actualNodo.sig
    
    print("None")
    




h = Nodo(1)
h.sig = Nodo(2)
h.sig.sig = Nodo(3)
h.sig.sig.sig = Nodo(4)
imprimirLista(h)



```
