# **Listas** 

Las listas simplemente enlazadas
- Cada nodo apunta solo al siguinte 
- El recorrido es unidireccional 
- El ultimo nodo a punta a None

## Conceptos básicos

### Lista
Estructura de datos compuesta por nodos enlazados.

### Nodo
Un nodo contiene:
- **Dato**
- **Referencias**
  - **Lista simple (LS)**: siguiente
  - **Lista doble (LD)**: siguiente, previo

## Operaciones
Las operaciones más comunes sobre listas son:
- Insertar (inicio, final)
- Usar
- Limpiar
- Imprimir
