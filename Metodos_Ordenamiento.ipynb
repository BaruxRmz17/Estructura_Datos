# Metodos de ordenamineto 
Los métodos de ordenamiento son algoritmos utilizados para organizar elementos en una lista o arreglo según un criterio específico, como el orden numérico o alfabético. A continuación, se describen algunos de los métodos de ordenamiento más comunes:

- Metodo de la burbuja
- Metodo de seleccion (sort)

```python
"""
Metodo de la burbuja
Ejercicio 1
"""
def burbuja(arr):
    cont = 0
    n = len(arr)
    for i in range(n):
        for j in range(n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                cont += 1
                
    return arr,cont

arre=[5,3,7,2,1,0,9,6,4,8,10,11,-1]
print(burbuja(arre))


                
```

```python
"""Metodo Selection Sort
Ejercicio 2
"""

def sort(arr):
    n = len(arr)
    for i in range(n):
        min = i
        for j in range(i+1, n):
            if arr[j] < arr[min]:
                min = j
        arr[i],arr[min] = arr[min],arr[i]
    
    return arr

arre=[5,3,7,200,1,0,37,9,6,4,8,10,11]
print(sort(arre))


    
```
