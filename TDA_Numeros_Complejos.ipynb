```python
class NumeroComplejo:
    # 'a' es la parte real y 'b' es la parte imaginaria
    def __init__(self,a,b):
        self.a = a
        self.b = b

    # Suma de numeros complejos
    def sumar(self, otroComplejo):
        return NumeroComplejo(self.a + otroComplejo.a, self.b + otroComplejo.b)

    # Multiplicacion de numeros complejos
    def multiplicar(self, otroComplejo):
        real = self.a * otroComplejo.a - self.b * otroComplejo.b
        imag = self.a * otroComplejo.b + self.b * otroComplejo.a
        return NumeroComplejo(real, imag)

    #modulo de un numero complejo
    def modulo(self):
        return(self.a**2 + self.b**2)**0.5
    
```

```python
z1 = NumeroComplejo(2,3)
z2 = NumeroComplejo(1,-1)

z3 = z1.sumar(z2)  # Suma de z1 y z2
z4 = z1.multiplicar(z2)  # Multiplicacion de z1 y z2

print(z3.a, z3.b)  # Imprime la suma: 3 2
print(z4.a, z4.b)  # Imprime la multiplicacion: 5
```
