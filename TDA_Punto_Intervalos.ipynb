# Ejercicios de punto a Punto (. -> .)

```python
import math
class Punto_Plano:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def distancia(self, otroPunto):
        return math.sqrt((self.x - otroPunto.x)**2 + (self.y - otroPunto.y)**2)


p1 = Punto_Plano(3,5)
p2 = Punto_Plano(2,3)

print(p1.distancia(p2))  # Imprime la distancia entre p1 y p2: 5.0

```

# Factorial

```python
def factorial(n):
    res = 1
    for i in range(n,1,-1):
        res = res * i
    return res

n = 5
print(factorial(n))  # Imprime el factorial de n: 120|
```

# Factorial de forma recursiva

```python
def factorialFormaRecursiva(n):
    if n == 1:
        return 1
    else:
        return n * factorialFormaRecursiva(n-1)
    
n = 5
print(factorialFormaRecursiva(n))  # Imprime el factorial de n:
```
