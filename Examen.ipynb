```python
#sumar recursivamente los elementos en posiciones pares de un arreglo 

def sumar_posiciones_pares(lista):
    if not lista:
        return 0
    elif len(lista) == 1:
        return lista[0]
    else:
        return lista[0] + sumar_posiciones_pares(lista[2:])
        
print(sumar_posiciones_pares([4,1,7,3,2]))
lista = [4,1,7,3,2] 
print(lista[2:])
print(lista[2:])



```

```python
#Crear una funcion recursiva que suma los elementos pares menos la suma de los elementos impares de un arreglo
def restar_pares(lista):
    if not lista:
        return 0
    
    elif len(lista) == 1:
        return lista[0] 
    else:
        return lista[0] - restar_pares(lista[2:])

print(restar_pares([3,7,9,4]))
```

```python
""" (U,D,L,R)
EJ: ['U','R','D','L']
output -> te extraño""" 

def arr_palabras(movi):
    x = 0
    y = 0

    for movi in movi:
        if movi == 'U':
            x += 1

        elif movi == 'R':
            y += 1

        elif movi == 'L':
            y -= 1

        else:
            x -= 1

    if x == 0 and y == 0:
        print("Regresaste")
    else: 
        print("te extraño")


# Pruebasd
arr_palabras(['U', 'R','D','D'])  # TE EXTRAÑO
arr_palabras(['U', 'D'])             # REGRESASTE
arr_palabras(['U', 'U', 'D', 'D'])   # REGRESASTE
        
```
