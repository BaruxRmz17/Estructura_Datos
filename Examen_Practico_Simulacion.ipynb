```python
num = 3

def eliminar_numero(lista):
    if len(lista) == 0:
        return []

    resto = eliminar_numero(lista[1:])

    if lista[0] == num:
        return resto
    else:
        return [lista[0]] + resto


a = [1, 2, 3, 3, 3, 4, 5]
print(eliminar_numero(a))
```
