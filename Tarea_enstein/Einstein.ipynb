#Bibliotecas

```python
from PIL import Image
import numpy as np
```

#Leer imagen

```python
def readImgPGM(path):
  # Open the PGM file
  img = Image.open(path)
  # Convert the image to a NumPy array
  img_array = np.array(img)

  # Display image information
  print(f"Image format: {img.format}")
  print(f"Image mode: {img.mode}") # L for grayscale
  print(f"Image size (width, height): {img.size}")

  return img_array
```

#Imprimir Imagen

```python
def printImg(img_array):
  # Convert the NumPy array back to a PIL Image object
  modified_img = Image.fromarray(img_array.astype('uint8'), 'L')

  # Save the modified image
  output_filename = "modified_output.pgm"
  modified_img.save(output_filename)
  print(f"\nModified image saved as {output_filename}")
```
