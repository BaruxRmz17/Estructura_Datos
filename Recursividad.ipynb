### **Ejercicio 1**

```python
def sumar(n):
    if n == 0:
        return 0
    else:
        return n + sumar(n - 1)

n = 5
print(sumar(n))     
```

### **Ejercicio 2**

```python
def numero_digitos(n):
    if n < 10:
        return 1
    else:
        return 1 + numero_digitos(n // 10)
    
n = 20 
print(numero_digitos(n))
```

```python
def potencia(base, exponente):
    if exponente == 0:
        return 1
    else:
        return base * potencia(base, exponente - 1)

print(potencia(2, 3)) 
```

```python
def max_lista(lista):
    if len(lista) == 1:
        return lista[0]
    else:
        return max(lista[0], max_lista(lista[1:]))

print(max_lista([3, 1, 4, 1, 5, 9, 2, 6, 10]))
```

```python
def suma_lista(lista):
    if len(lista) == 1:
        return lista[0]
    else:
        return sum(lista)
    
print(suma_lista([1,2,3,4,5,6]))
```

```python
def palindroma(palabra, inicio, fin=None):
    if fin == None:
        fin = len(palabra) - 1
    if inicio >= fin:
        return True
    if palabra[inicio] != palabra[fin]:
        return False
    return palindroma(palabra, inicio + 1, fin - 1)

print(palindroma("oso", 0))
```

```python
def fibonacci(n):
    # casos base
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
def fibonacci(n):
    # casos base
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


# pedir el número n
n = int(input("Ingresa un número n: "))
print(f"{fibonacci(n)}")


```

```python
def apariciones(lista, x, i = 0):
    if i == len(lista):
        return 0
    if lista[i] == x:
        return 1 + apariciones(lista, x, i + 1)
    return apariciones(lista, x, i + 1)

print(apariciones([1,2,3,4,5,1,1,1], 1))
        
```

```python
def MCD(a,b):
    if b == 0:
        return a
    else:
        return MCD(b, a % b)
    
print(MCD(48,18))
```

```python
def invertir_numero(n, resultado=0):
    if n == 0:
        return resultado
    else:
        return invertir_numero(n // 10, resultado * 10 + n % 10)

print(invertir_numero(123456789))
```
