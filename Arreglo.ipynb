# **Arreglos**

Es una coleccion de datos de cualquier tipo pude haber desde una dimension en adelante aunque mas de dos dimensiones ya se considera una matriz. En `python` no se necesita especificar el tipo de dato que tendra en comparacion con otros lenguajes como c++, java etc.

>Una estructura de datos es una forma de organizar y acomdarlos.

*Es una coleccion ordenada de elementos del mismo tipo, accesibles mediante un unico indice*
- indices consecutivos
-  acceso en tiempo constante(o) 

>En  `python ` los arreglos unidimensionales se representan comunmente con listas 

```
edades = [18,20,11,22]
print(edades[0])
```

#### **Operaciones basicas**
- Acceso a un elemento
- Modificacion de un elemento
- Eliminacion 
- Recorrido


- En las Matrices el primer indice es la fila 


````python
#Ejemplo 1
numeros = [4,7,2,9,99]
suma = 0

for x in numeros:
    suma += x

print(suma)
````

````python
# ejemplos 2
matriz = [
    [1,2,3],
    [4,5,6]]

for fila in matriz:
    for elemento in fila:
        print(elemento , end= "")
    print()
````

````python
# ejemplo 3
# calcular las suma de una matriz
matriz = [[1,2] , [3,4]]
suma = 0

for fila in matriz:
    for elemento in fila:
        suma += elemento
print(suma)
````

````python
# Ejemplo 4
def busqueda_linea(arreglo , valor):
    for i in range(len(arreglo)):
        if arreglo[i] == valor:
            return i
    return -1
print(busqueda_linea([4,7,2,9] , 2))
````

````python
#Encontrar el valor maximo en un arreglo

valores = [8,3,10,2]
maximo = valores[0]

for x in valores:
    if x > maximo:
        maximo = x

print(maximo)
````

### **Ejercicio 1**

````python
def promedio(arr):
    suma = 0
    for x in arr:
        suma += x
    return suma / len(arr)

print(promedio([4,7,2,9]))
````

````python
def pares(ep):
    ele = 0
    for i in ep:
        if i % 2 == 0:
            ele += 1
    return ele
ep = [6,2,4,5]
print(pares(ep))
````

### **Ejercicio 3**

````python
def invertir(arreglo):
    inv = []
    for i in range(len(arreglo)-1,0,-1):
        inv.append(arreglo[i])
    return inv  

print(invertir([1,3,4,5]))
    
````

### **Ejercicio 5**

````python
def eliminar_duplicado(arreglo):
    sin_duplicado=[]
    for i in arreglo:
        if i not in sin_duplicado:
            sin_duplicado.append(i)
    return sin_duplicado

print(eliminar_duplicado([1,2,2,3,4,4,5]))
````

### **Suma de una matriz**

````python
#Suma de las filas de una matriz
def suma_filas_matriz(matriz):
    resultado = []

    for fila in matriz:
        suma = 0
        for elemento in fila:
            suma += elemento

        resultado.append(suma)

    return resultado


print(suma_filas_matriz([[1,2,3],[4,5,6]]))
````

````python
#Suma de las columnas de una matriz
def suma_columnas(matriz):
    columnas = len(matriz[0])
    resultado = []

    for c in range(columnas):
        suma = 0
        for f in range(len(matriz)):
            suma += matriz[f][c]
        resultado.append(suma)

    return resultado

print(suma_columnas([[1,2,3],[4,5,6]]))

````

````python
#Determinar si una matriz es cuadrada
def _es_cuadrada(matriz):
    filas = len(matriz)
    for fila in matriz:
        if len(fila) != filas:
            return False
    return True

print(_es_cuadrada([[1,2,3],[4,5,6]]))
````

````python
#Obtener la diagonal principal de una matriz 
d = []

def diagonal_principal(matriz):
    for i in range(len(matriz)):
        d.append(matriz[i][i])
    return d
        

matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(diagonal_principal(matriz))
````

````python
#Transponer una matriz 
def transponer_matriz(matriz):
    t = []
    for i in range(len(matriz[0])):
        t.append([])
        for j in range(len(matriz)):
            t[i].append(matriz[j][i])
    return t

matriz = [
    [1, 2, 3],
    [4, 5, 6], 
    [7, 8, 9]]

print(transponer_matriz(matriz))
````
